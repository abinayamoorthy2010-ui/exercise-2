```python
class Student:
    def __init__(self, name, rollNumber):
        self.name = name
        self.rollNumber = rollNumber
    def display(self):
        print("Name:", self.name)
        print("Roll Number:", self.rollNumber)
class UndergraduateStudent(Student):
    def __init__(self, name, rollNumber, r1, r2 ,r3):
        super().__init__(name, rollNumber)
        self.r1 = r1
        self.r2 = r2
        self.r3 = r3
    def calculateAverage(self):
        return (self.r1 + self.r2 + self.r3) /3
    def display(self):
        print("\n--- Undergraduate Student ---")
        super().display()
        print("Average Marks:", self.calculateAverage())
class PostgraduateStudent(Student):
    def __init__(self, name, rollNumber, r1, r2, r3, r4):
        super().__init__(name, rollNumber)
        self.r1 = r1
        self.r2 = r2
        self.r3 = r3
        self.r4 = r4
    def calculateAverage(self):
        return (self.r1 + self.r2 + self.r3 + self.r4) / 4
    def display(self):
        print("\n--- Postgraduate Student ---")
        super().display()
        print("Average Marks:", self.calculateAverage())
if __name__ == "__main__":
    ug = UndergraduateStudent("Atchaya", 1,90,85,97)
    pg = PostgraduateStudent("Abi", 2, 70, 88, 90, 84)
    ug.display()
    pg.display()
```
